# 1. 센서데이터 기기별 병합

```python
import pandas as pd
import numpy as np
import os
import datetime
import warnings

warnings.filterwarnings(action = "ignore")
```

```python
dp_columns = ["Time","Location","DP"]
th_columns = ["Time","Location","Temperature","Humidity"]
light_expose_columns = ["Location","Time","over_0.3_delta","over_0.5_delta","over_1.0_delta","over_5.0_delta","over_0.3_sigma","over_0.5_sigma","over_1.0_sigma"]
```

```python
dp_01_2 = pd.DataFrame(columns = dp_columns)
th_01_0 = pd.DataFrame(columns = th_columns)
th_02_1 = pd.DataFrame(columns = th_columns)
light_expose_1 = pd.DataFrame(columns = light_expose_columns)
light_expose_2 = pd.DataFrame(columns = light_expose_columns)
light_expose_4 = pd.DataFrame(columns = light_expose_columns)
light_expose_5 = pd.DataFrame(columns = light_expose_columns)
```

* 날짜는 2020.04.27 ~ 2020.08.11 까지

```python
day_list = []
start = datetime.datetime.strptime("2020.04.27","%Y.%m.%d")
end = datetime.datetime.strptime("2020.08.12","%Y.%m.%d")
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]
for date in date_generated:
    day_list.append(date.strftime("%Y%m%d"))
```

```python
root = "./data/"
file_heads_df = {"DP01_2_":dp_01_2,"TH01_0_":th_01_0,"TH02_1_":th_02_1,"노광1호기130_":light_expose_1,"노광2호기131_":light_expose_2,"노광4호기132_":light_expose_4,"노광5호기133_":light_expose_5}
```

```python
empt_list = []
for day in day_list:
    for file_head in file_heads_df:
        path = root + day + "/" + file_head + day + ".csv"
        try:
            tmp_df = pd.read_csv(path,skiprows = 1,encoding = 'euc-kr', header = None)
            tmp_df = tmp_df.dropna(axis = 1)
            if file_head == "DP01_2_":
                tmp_df.columns = dp_columns
            elif file_head in ["TH01_0_","TH02_1_"]:
                tmp_df.columns = th_columns
            else:
                tmp_df.columns = light_expose_columns
            file_heads_df[file_head] = pd.concat([file_heads_df[file_head],tmp_df])
            print(file_heads_df[file_head].shape)
        except:
            emp_list = empt_list.append(str(file_head) + str(day))
            pass
```

```python
dp_01_2 = file_heads_df["DP01_2_"]
th_01_0 = file_heads_df["TH01_0_"]
th_02_1 = file_heads_df["TH02_1_"]
light_expose_1 = file_heads_df["노광1호기130_"]
light_expose_2 = file_heads_df["노광2호기131_"]
light_expose_4 = file_heads_df["노광4호기132_"]
light_expose_5 = file_heads_df["노광5호기133_"]
```

```python
print(dp_01_2.shape)
print(th_01_0.shape)
print(th_02_1.shape)
print(light_expose_1.shape)
print(light_expose_2.shape)
print(light_expose_4.shape)
print(light_expose_5.shape)
```

```python
def strip_date(df, my_format = "%Y-%m-%d %H:%M:%S"):
    df["Time"] = pd.to_datetime(df["Time"],format = my_format)

    df["month"] = list(map(lambda x: x.month,df["Time"]))
    df["day"] = list(map(lambda x: x.day,df["Time"]))
    df["weekday"] = list(map(lambda x: x.weekday(),df["Time"]))
    df["hour"] = list(map(lambda x: x.hour,df["Time"]))
    df["minute"] = list(map(lambda x: x.minute,df["Time"]))
    df["second"] = list(map(lambda x: x.second,df["Time"]))
    return df
```

```python
dp_01_2 = strip_date(dp_01_2)
th_01_0 = strip_date(th_01_0)
th_02_1 = strip_date(th_02_1)
light_expose_1 = strip_date(light_expose_1,my_format = "%Y%m%d%H%M%S")
light_expose_2 = strip_date(light_expose_2,my_format = "%Y%m%d%H%M%S")
light_expose_4 = strip_date(light_expose_4,my_format = "%Y%m%d%H%M%S")
light_expose_5 = strip_date(light_expose_5,my_format = "%Y%m%d%H%M%S")
```

```python
file_names = list(file_heads_df.keys())
save_root_path = "./preprocessed/"
save_path = list(map(lambda x: save_root_path + x + ".csv", file_names))
df_list = [dp_01_2,th_01_0,th_02_1,light_expose_1,light_expose_2,light_expose_4,light_expose_5]

for i in range(len(save_path)):
    df_list[i].to_csv(save_path[i])
```

```python
par_path = os.curdir + "/data/Warning/"
```

```python
warning_list = os.listdir("./data/Warning/")
warning_df = pd.DataFrame(columns = light_expose_columns)
```

```python
for warning in warning_list:
    path = par_path + warning
    tmp_df = pd.read_csv(path,encoding = "euc-kr",header = None,skiprows = 1)
    tmp_df = tmp_df.dropna(axis = 1)
    tmp_df.columns = light_expose_columns
    warning_df = pd.concat([warning_df,tmp_df])
```
