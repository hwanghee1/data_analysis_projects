```python
def merge_dates(x):
    date_time = datetime.datetime(2020,int(x["month"]),int(x["day"]),int(x["hour"]),int(x["minute"]))
    return date_time
```

```python
def merge_columns(tmp_df,date_columns = ["month","day","hour","minute"]):
    time_list = tmp_df[date_columns].apply(lambda x: merge_dates(x), axis = 1)
    tmp_df["Time"] = time_list
    tmp_df = tmp_df.drop(date_columns,axis = 1)
    return tmp_df
```

```python
def divide_un_over(processed_df,date_time):
    under_df = processed_df.loc[np.where(processed_df["Time"] < date_time)]
    over_df = processed_df.loc[np.where(processed_df["Time"] > date_time)]
    return under_df, over_df
```

```python
data_columns = ["over_0.3_delta","over_0.5_delta","over_1.0_delta","over_5.0_delta","over_0.3_sigma","over_0.5_sigma","over_1.0_sigma"]
```

* df_preprocesse :
    - 노광기 파티클 데이터 중 warning 데이터에 포함된 행에 대해 warning표시가 있으면 1, 없으면 0으로 만드는 열을 생성해주는 코드

```python
def df_preprocess(light_df,warn_df):
    light_df["warning"] = list(map(lambda x : int(x in warn_df["Time"].values),light_df["Time"]))
    tmp_df = light_df.groupby(["month","day","hour","minute"])[data_columns + ["warning"]].max()
    merged_tmp_df = merge_columns(tmp_df.reset_index()) 
    return merged_tmp_df
```

```python
processed_1 = df_preprocess(light_expose_1,warn_1)
processed_2 = df_preprocess(light_expose_2,warn_2)
processed_4 = df_preprocess(light_expose_4,warn_4)
processed_5 = df_preprocess(light_expose_5,warn_5)
```

* make_warn_point : 
    - warning 발생 시점을 그래프에 그리기 위한 코드

```python
def make_warn_point(processed_df):
    warn_point = processed_df[processed_df["warning"] != 0]
    return warn_point
```

* divide_un_over : 어느 시점 기준 전, 후 데이터로 분할하는 함수

## 노광 1호 

```python
def plot_line_graph(data_frame):
    warn_point = make_warn_point(data_frame)
    
    for column in data_columns:
        plt.figure(figsize = (16,6))
        sns.lineplot(x = "Time",y = column,data = data_frame)
        plt.plot(warn_point["Time"],warn_point[column],'ro')
        plt.title("분당 " + column +" 최대값", size = 18)
        plt.xticks(rotation = 45,size = 12)
        plt.xlabel("Time",size = 14)
        plt.ylabel("max of " + column,size = 14)
        plt.show()
```

* plot 제목 설명  
    ex) 분당 over_0.3_delta 최대값
    - 분당, 최대값 : 데이터를 분단위로 결합하여 1분 내 over_0.3_delta 최대값을 뽑았음
    - over_0.3_delta : &#62;0.3μmΔ (0.3μm < 파티클 수 < 0.5μm)  
    - over_0.3_sigma : &#62;0.3μmΣ (0.3μm < 파티클 수)
